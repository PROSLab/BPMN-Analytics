```python
import numpy as np
import os
import csv
import time
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('process.csv', low_memory=False, delimiter=";", encoding = "ISO-8859-1", index_col=False)
csvfilename = "process"
np.seterr(divide='ignore', invalid='ignore')
filename = csvfilename+"PearsonCorrelation.csv" 
header = df.columns.values.tolist()
header.insert(0, '\\' )
with open(filename, 'a', newline='') as csvfile:
            csvwriter = csv.writer(csvfile)
            csvwriter.writerow(header)           
for index, element in enumerate(df):
    rho_element_list = []
    x_df = np.array(df.loc[0:, element])    
    for index2, element2 in enumerate(df):
        y_df = np.array(df.loc[0:, element2])
        my_rho = np.corrcoef(x_df,y_df)
        currentRho = my_rho[1][0]        
        rho_element_list.append(currentRho)
    with open(filename, 'a', newline='') as csvfile:
        csvwriter = csv.writer(csvfile)
        csvwriter.writerow([element] + rho_element_list)
```
